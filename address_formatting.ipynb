```python
address = pd.read_csv('addresses.csv', dtype={'zipcode': str})

display(address.head(10))
```

```python
display(address[address['province'] == 'ชุมพร'])
display(address[address['province'] == 'ชุมพร'].shape)
```

```python
address = address[address['zipcode'].notna()]
display(address[address['province'] == 'ชุมพร'])


```

```python
address['province_en'] = address['province_en'].replace('buogkan', 'Bueng Kan') 
display(address[address['province'] == 'บึงกาฬ'])
```

```python
print(address.amphoe_code.isna().sum())
display(address[address.amphoe_code.isna()])
```

```python
address.loc[address['zipcode'].isna()]
```

```python
address_json = dict()

for idx in address.index:

    district_th = address.district[idx]
    amphoe_th = address.amphoe[idx]
    province_th = address.province[idx]
    
    district_en = address.district_en[idx]
    amphoe_en = address.amphoe_en[idx]
    province_en = address.province_en[idx]
    zipcode = address.zipcode[idx]

    if province_th not in address_json:
        province_obj = {'value': province_th, 'en': province_en, 'th': province_th, 'nextValue': dict()}
        address_json[province_th] = province_obj

    if amphoe_th not in address_json[province_th]['nextValue']:
        amphoe_obj = {'value': amphoe_th, 'en': amphoe_en, 'th': amphoe_th, 'nextValue': dict()}
        address_json[province_th]['nextValue'][amphoe_th] = amphoe_obj
    
    
    district_obj = {'value': district_th, 'en': district_en, 'th': district_th, 'nextValue': str(int(zipcode))}
    address_json[province_th]['nextValue'][amphoe_th]['nextValue'][district_th] = district_obj
```

```python
address_json
```

```python
with open("addresses.json", "w") as outfile:
    outfile.write(str(address_json))
```
